```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
from glob import glob
from skimage.color import rgb2gray
```

```python
train_files = np.array(glob("train/*/*"))
```

```python
len(train_files)
```

```python
train_files[0]
```

```python
img = cv2.imread(train_files[0])
```

```python
plt.imshow(img)
```

```python
# apply gray scale filter
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray,cmap='gray')
```

```python
# apply threshold filter
thresh, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
```

```python
plt.imshow(binary,cmap='gray')
```

```python
# Declare the variables we are going to use
ddepth = cv2.CV_16S
kernel_size = 3
window_name = "Laplace Demo"
```

```python
 # Remove noise by blurring with a Gaussian filter
src = cv2.GaussianBlur(binary, (3, 3), 0)
```

```python
plt.imshow(src,cmap='gray')
```

```python
# Apply Laplace function
dst = cv2.Laplacian(src, ddepth, ksize=kernel_size)
```

```python
plt.imshow(dst,cmap='gray')
```

```python
# converting back to uint8
abs_dst = cv2.convertScaleAbs(dst)
```

```python
plt.imshow(abs_dst,cmap='gray')
```
