```python
#Converting image into grayscale using opencv
import cv2 
import numpy as np
import numpy as np
from skimage.io import imshow, imread
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
path = r'/content/drive/MyDrive/project_data/train/DME/DME-15307-1.jpeg'
src = cv2.imread(path)
#image = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)

#Applying 7x7 gaussian blur
image = cv2.GaussianBlur(gray, (7, 7), 0)
  
# Displaying the image 
cv2_imshow(image)
```

```python
#Applying threshold in image
th, image = cv2.threshold(src,127,255, cv2.THRESH_TRUNC);

cv2_imshow(image)
```

```python
#Applying erosion in our image
kernel = np.ones((5,5),np.uint8)
image = cv2.dilate(image,kernel,iterations = 1)
cv2_imshow(image)
```

```python
#Applying simple blob detector
detector = cv2.SimpleBlobDetector_create()
keypoints = detector.detect(image)
blank = np.zeros((1,1))
image = cv2.drawKeypoints(image, keypoints, blank, (0,255,255), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
cv2_imshow(image)
```
